**ENTER YOUR RESPONSE HERE.**

To fit the model, we would need roughly half of the dataset to fir the model. This means that we would need 4772 observations to train the model and fit it. The next half of the dataset would be to predict the status and pressure and then run some metrics such as MSE or RMSE.

b) Write code to fit the model and compute the predicted value for an extratropical storm with a pressure of 970. Your final answer should be a single number, but make sure your code clearly contains all the necessary steps to arrive at the appropriate number.

```python
# ENTER ANY CODE HERE
storms_train = df_storms.loc[:4772].copy()
storms_test = df_storms.loc[4773:].copy()

X_train = storms_train[["status", "pressure"]]
X_test = storms_test[["status", "pressure"]]
y_train = storms_train["tropicalstorm_force_diameter"]

from sklearn.linear_model import LinearRegression
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.pipeline import make_pipeline

ct = make_column_transformer(
    (OneHotEncoder(), ["status"]),
    remainder="passthrough"
)
```

```python
pipeline = make_pipeline(
    ct,
    LinearRegression()
)

pipeline.fit(
    X=X_train[["status", "pressure"]],
    y=y_train
)
```

```python
x_test = pd.Series()
x_test["status"] = "extratropical"
x_test["pressure"] = 970

x_test = pd.DataFrame([x_test])
```

```python
pipeline.predict(x_test)[0]
```

**ENTER YOUR RESPONSE HERE.**

For a extratropical storm and pressure of 970, my model predicts the force diameter to be 392.

c) Write code to compute a number which measures the effectiveness of the model from the previous part. Your final answer should be a single number, but make sure your code clearly contains all the necessary steps to arrive at an appropriate number. **Write a clearly worded sentence interpreting the value in context.**

```python
# ENTER ANY CODE HERE

from sklearn.model_selection import cross_val_score
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np

cv_scores = cross_val_score(pipeline, df_storms[["status", "pressure"]], df_storms["tropicalstorm_force_diameter"], cv=10, scoring='neg_mean_squared_error')
cv_rmse = np.sqrt(-cv_scores)
print(f"Cross-validated RMSE: {cv_rmse.mean()}")
```
