```python
import pandas as pd

df_earnings_height = pd.read_csv('https://raw.githubusercontent.com/kevindavisross/data301/main/data/Earnings_and_Height.csv')
df_earnings_height
```

1\. height

```python
# YOUR CODE HERE
df_height = df_earnings_height.height
df_height.plot.hist()
```

```python
df_height.mean()
```

```python
df_height.median()
```

```python
# Standard deviation, mean absolute deviation, and interquartile range
print(df_height.std())

print((df_height - df_height.mean()).abs().mean())

print(df_height.quantile(0.75) - df_height.quantile(0.25))
```

**YOUR RESPONSE HERE.**

The mean and median are about the same so I would say its a symmetric distribution. The standard deviation is 3.9 inches and the mean absolute deviation is 3.2 inches, which means that the data is on average 3.9 inches away from the mean. The interquartile range is 5.5 inches, which means that the middle 50% of the data is within 6 inches of each other.

2\. years of education. What do you think is most appropriate: a histogram or a bar chart? Why?
- Histogram because it is a quantitative variable

```python
# YOUR CODE HERE
df_education = df_earnings_height.educ
df_education.plot.hist()
```

```python
print(df_education.mean())
print(df_education.median())
print(df_education.std())
print((df_education - df_education.mean()).abs().mean())
print(df_education.quantile(0.75) - df_education.quantile(0.25))
```

**YOUR RESPONSE HERE.**

This distribution is skewed right since the mean is greater than the median. The standard deviation is 2.6 which means that the data is on average 2.6 years away from the mean. Finally the IQR is 4 years which means that the middle 50% of the data is within 4 years of each other.
