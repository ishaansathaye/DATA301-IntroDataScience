## Recreations

Use Plotly or Altair to recreate some of the graphics we have already produced using Pandas

The colored scatterplot from the penguins data

```python
alt.Chart(df_penguins).mark_point().encode(
    x="bill_length_mm",
    y="bill_depth_mm",
    color="species"
)
```

The faceted scatterplot from the penguins data


```python
alt.Chart(df_penguins).mark_point().encode(
    x="bill_length_mm",
    y="bill_depth_mm",
    facet="species"
)
```

The side-by-side bar plot from the COVID data

```python
df_country = df_covid.groupby(["country", "age"])["fatality"].mean().reset_index()

# Altair Version 5
alt.Chart(df_country).mark_bar().encode(
    x="age",
    y="fatality",
    color="country",
    xOffset="country"
)

# Altair Version 4
# alt.Chart(df_country).mark_bar().encode(
#     x="country",
#     y='fatality',
#     color="country",
#     column="age"
# )
```

The overlaid histograms (or density plots) of sale price by building type for the Ames housing data

```python
# overlaid histograms in altair
# alt.Chart(df_ames).mark_bar().encode(
#     x="SalePrice",
#     y="count()"
# )
# df_ames.columns
df_building = df_ames.groupby(["Bldg Type"])["SalePrice"].value_counts().reset_index(name="count")
df_building

alt.Chart(df_building).transform_fold( 
    ['Bldg Type'], 
    as_=['Columns', 'Values'] 
).mark_area( 
    opacity=0.5, 
    interpolate='step'
).encode( 
    alt.X('SalePrice', bin=alt.Bin(maxbins=100)), 
    alt.Y('count(SalePrice)'), 
    alt.Color('Bldg Type:N')
)
```

You should be able to see the progression of your graph across the several drafts, and not just the end result.
